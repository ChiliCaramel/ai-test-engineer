一、可变数据类型；不可变数据类型
 可变：list dict  - 内存地址不变，但是数据内容可变
 不可变：int float str bool tuple

 有序的数据类型：list tuple dict


```python
# 可变数据类型：数据修改后，地址不变（同一块内存地址）
# list 论证

my_list = [1,2,3]
print(f"my_list id is {id(my_list)}") # my_list id is 4817186688

my_list.append(4)
print(f"my_list id is {id(my_list)}") # my_list id is 4817186688
```

```python
# dict 论证

my_dict = {"name":"Cara","id":888} # my_dict id is 4817185984
print(f"my_dict id is {id(my_dict)}")

my_dict["uid"] = 666
print(f"my_dict id is {id(my_dict)}") # my_dict id is 4817185984
```

```python
# set
my_set = {1,2,3}
print(f"my_set id is {id(my_set)}") # my_set id is 4726389696


my_set.add(4)
print(f"Now my_set id is {id(my_set)}") # Now my_set id is 4726389696
```

```python
# 不可变数据类型：数据修改后，内存地址会变。

# int
a = 10
print(f"a id is {id(a)}") # a id is 4398376728

# 重新赋值
a = 20
print(f"Now a id is {id(a)}") # Now a id is 4398377048
```

```python
# float

b = 1.0
print(f"b id is {id(b)}") # b id is 4708557040

# 重新赋值
b = 3.0
print(f"Now b id is {id(b)}") # Now b id is 4726644080
```

```python
# str
s = "Hello"
print(f"s id is {id(s)}") # s id is 4726560240

s = "Hi"
print(f"Now s id is {id(s)}") # Now s id is 4726558832
```

```python
# tuple
t = (1,2,3)
print(f"t id is {id(t)}") # t id is 4817143360

t = (4,5,6)
print(f"Now t id is {id(t)}") # Now t id is 4818106176

t[0] = 99 # TypeError: 'tuple' object does not support item assignment

```

```python
# bool
x = True
print(f"x id is {id(x)}") # x id is 4397370728


x = False
print(f"Now x id is {id(x)}") # Now x id is 4397371456
```

```python
# 定义一个列表变量，包含几个元素，尝试对列表进行添加、删除和修改操作，观察每次操作后列表的内容和内存地址是否发生变化。

my_list = [1,2,3,4]
print(f"my_list id is {id(my_list)}") # my_list id is 4817724160

my_list.append(5)
print(f"Now my_list:{my_list} id is {id(my_list)}") # Now my_list:[1, 2, 3, 4, 5] id is 4817724160

my_list.remove(3)
print(f"Now my_list:{my_list} id is {id(my_list)}") # Now my_list:[1, 2, 4, 5] id is 4817724160

my_list.insert(1,"a")
print(f"Now my_list:{my_list} id is {id(my_list)}") # Now my_list:[1, 'a', 2, 4, 5] id is 4817724160
```

```python
# 定义一个元组变量，包含与列表相同的元素，尝试对元组进行类似的操作，观察会发生什么。

t = (1,2,3,4)
print(f"t id is {id(t)}")  # t id is 4818039520

# 元组是不可变类型，所以没有支持自身修改的方法。
# 如果要进行 增 改 删 变量内元素的操作，只能通过 - 
# 拼接：
t = t + (5,)
print(f"Now t:{t} id is {id(t)}") # Now t:(1, 2, 3, 4, 5) id is 4818042800

# 切片：
t = t[1:]
print(f"Now t:{t} id is {id(t)}") # Now t:(2, 3, 4, 5) id is 4818039520

# 内存地址发生变化
```

```python
# 定义一个字典变量，包含几个键值对，尝试对字典进行添加、删除和修改操作，观察每次操作后字典的内容和内存地址是否发生变化。

my_dict = {"a":1,"b":2,"c":3}
print(f"my_dict id:{my_dict} is {id(my_dict)}") # my_dict id:{'a': 1, 'b': 2, 'c': 3} is 4817776384

my_dict["d"] = 4
print(f"Now my_dict:{my_dict} id is {id(my_dict)}") # Now my_dict:{'a': 1, 'b': 2, 'c': 3, 'd': 4} id is 4817776384

my_dict.pop("c")
print(f"Now my_dict:{my_dict} id is {id(my_dict)}") # Now my_dict:{'a': 1, 'b': 2, 'd': 4} id is 4817776384

my_dict["b"] = 22
print(f"Now my_dict:{my_dict} id is {id(my_dict)}") # Now my_dict:{'a': 1, 'b': 22, 'd': 4} id is 4817776384

# 内存地址不变
```

```python
# 编写一个函数，接受一个整数参数，并返回该整数的平方。
def square(num:int) -> int:
    sq = num ** 2
    return sq

print(square(2))
```

```python
# 编写一个函数，接受两个参数，一个整数和一个字符串，返回一个字符串，描述整数参数的值和字符串参数的长度
def describe_params(num:int,s:str):

    len_s = len(s)
    return f"整数参数为{num},字符串参数的长度为{len_s}"

print(describe_params(22,"dsfsf"))
```

```python
# 编写一个函数，使用动态参数*args，接受任意数量的整数，并返回这些整数的平均值
def calculate_avg(*args):
    s = 0
    l = len(args)
    if not l :
        return 0
    else:
        for num in args:
            s+=num
    
    return s/l

print(calculate_avg(1,2,3))
```

```python
# 编写一个函数，使用动态参数**kwargs，接受任意数量的关键字参数，并返回一个包含所有参数和值的字符串
def output_str(**kwargs):
    if not kwargs:
        return "Warning: No kwargs"
    try:
        formatted_items = [f"{k}:{v}" for k,v in kwargs.items()]
        res = ",".join(formatted_items)
        
        return res
    
    except Exception as e:
        return f"处理参数时发生错误: {e}"

print(output_str(a=1,b=2))
```

```python
# 尝试使用默认参数值编写一个函数，该函数接受一个字符串和一个可选的颜色参数（默认为"黑色"），并返回一个格式化的字符串，显示字符串内容和颜色。
def text_color(text:str,color = "黑色"):
    return f"输入的字符串为{text},颜色是{color}"

print(text_color("abc"))
print(text_color("abcd","红色"))
```

```python
# 编写一个函数，接受一个列表和一个函数作为参数，该函数会对列表中的每个元素应用提供的函数，并返回一个新的列表，包含应用函数后的结果
def deco(items,func):
    result = []

    for it in items:
        result.append(func(it))
    return result

def square(num):
    return num ** 2

print(deco([1,2,3],square))
```
