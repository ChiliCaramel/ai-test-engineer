```python
class Animals(object):

    def feet(self):
        print("动物的jio")
    
    def mouth(self):
        print("动物的嘴")


class Dog(Animals): # 继承父类Animals

    def fur(self):
        print("狗的毛")

    def feet(self):
        print("狗的jio")
    
    def mouth(self):
        print("狗的嘴")

class Cat(Animals):

    def feet(self):
        print("猫的jio")
    
    def mouth(self):
        print("猫的嘴")

# 多态
def run_func(obj):
    obj.feet()

dog = Dog()
cat = Cat()
animals = Animals()

run_func(dog)
run_func(cat)
run_func(animals)
```

继承 - 子类可以继承父类，方式是将父类名称写在子类名后的括号内

子类继承父类的属性和方法，支持子类重写父类方法：名称相同，参数相同，方法体不同
子类可以新增自己的属性和方法

多态 - 多态是指调用相同的方法，传入的对象不同，表现出不同的行为
实现的方法是，可以让多个类拥有同名方法，当某个方法统一调用这个方法时，会根据传入的对象执行对应的方法体。

```python
class Animal():
    
    def make_sound(self):
        print("Animal sounds.")

class Dog(Animal):

    def make_sound(self):
        print("Sounds of dog - Woof Woof Woof")


class Cat(Animal):

    def make_sound(saelf):
        print("Sounds of cat - Meow Meow Meow")

class AnimalSoundSimulator:

    def simulate_sound(self,obj):
        obj.make_sound()


if __name__ == "__main__":

    dog = Dog()
    cat = Cat()
    simulator = AnimalSoundSimulator()

    simulator.simulate_sound(dog)
    simulator.simulate_sound(cat)
```

总结：

1. Dog ， Cat 作为子类继承父类Animals 并重写父类方法 - make_sound()
   simulate_sound(self,obj) 统一调用make_sound()方法,由于传入的对象不同(Dog，Cat)，实现不同的行为（狗叫，猫叫） -- 即多态

2. 坑点 - python类中的实例方法一定记得要加 self 
因为python编译的过程如下：

simulator.simulate_sound(dog)
相当于：

AnimalSoundSimulator.simulate_sound(
    simulator,
    dog
)-> 需要有一个self 承接 simulator

否则会报错 - TypeError: AnimalSoundSimulator.simulate_sound() takes 1 positional argument but 2 were given.

和Java不同，Java是隐式的，this不用一直写编译器会自动处理，python是显示的self一定要写，作为第一个参数被接收。
