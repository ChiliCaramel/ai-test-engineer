# 常用的正则表达式
\d     # 数字 0-9
\w     # 字母数字下划线
\s     # 空白字符
+      # 一个或多个
*      # 零个或多个
?      # 零个或一个
{n}    # 精确n个
{n,m}  # n到m个
^      # 开头
$      # 结尾

不可变类型：int, float, str, bool, tuple
修改后地址会变，不存在拷贝问题

可变类型：list, dict, set
修改后地址不变，存在拷贝问题

```python
# 直接赋值：两个变量指向同一个内存地址
a = [1, 2, 3]
b = a
print(id(a))
print(id(b))      # b 和 a 指向同一个地址 2312873599936

a.append(4)
print(b)        # [1, 2, 3, 4] b 也被改了
```

```python
import copy

a = [1, 2, [3, 4]]

# 浅拷贝：只复制第一层
b = copy.copy(a)

# 地址
print(id(a))        # 不同地址
print(id(b))        # 不同地址
print(id(a[2]))     # 地址相同  2312873771264
print(id(b[2]))     # 地址相同  2312873771264

# 修改第一层a
a.append(5)
print(b)            # [1, 2, [3, 4]] 

# 修改内层列表 - a改 b也会跟着更改
a[2].append(5)
print(b)            # [1, 2, [3, 4, 5]] 
```

```python
import copy

a = [1, 2, [3, 4]]

# 深拷贝：所有层都复制，完全独立
b = copy.deepcopy(a)

# 验证地址
print(id(a))        # 地址不同 2312873721216
print(id(b))        # 地址不同 2312873721792
print(id(a[2]))     # 地址不同 - 内层地址也是互相独立的  2312873717248
print(id(b[2]))     # 地址不同 - 内层地址也是互相独立的  2312873800768

# 修改任何层都互不影响
a[2].append(5)   # 更改a的内层
print(b)            # [1, 2, [3, 4]] 
```
