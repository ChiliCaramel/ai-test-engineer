```python
import pymysql

class MysqlConnection():
    """
    类属性 实例属性的总结与练习
    """    
    # 类属性
    # 属于类，不属于实例，全局配置/共享数据 - global
    pwd = 123456
    port = 3306

    def __init__(self):

        # 实例属性
        self.ip = '127.0.0.1'
        self.user_account = 'root'
    
    def select(self):
        print("---select----")
    
    def change_ip(self):
        # 类属性 全局配置

        print(MysqlConnection.port)
        print("before:", vars(self))

        self.ip = "192.168.1.1"

        print("after:", vars(self))
    
db = MysqlConnection()
# 类属性可以通过实例访问 - db.pwd
print(f"before - ip is {db.ip},pwd is {db.pwd}")
print(f"before - {vars(db)}")

# 实例属性可以通过外部来修改 - dp.ip
db.ip = '00000'
db.pwd = '1111'

# 类属性不能通过实例修改，实例只是新创建了一个同名属性，并且基于python  实例属性优先级 > 类属性优先级 的机制，在本次实例回收or删除db.pwd前， 都只能访问到该属性
print(f"after - ip is {db.ip},pwd is {db.pwd}, class pwd {MysqlConnection.pwd}")
print(f"after - {vars(db)}")

print("----Now change class pwd----")
# 类的属性直接通过类来访问和修改是可以的
MysqlConnection.pwd = "aka"
del db.pwd
# 修改后，实例访问到的值也会跟着改变
print(db.pwd) # aka
print(MysqlConnection.pwd) # aka
print("-"*5)

# 实例属性可以通过类内部修改和访问
db.change_ip()

# 实例属性无法通过类的方式访问
# print(MysqlConnection.user_account) # AttributeError: type object 'MysqlConnection' has no attribute 'user_account'
```

```python
import pymysql

class MysqlConnectionPrivate():
    """
    类的私有属性 和 实例的私有属性的总结与练习
    """
    
    pwd = 123456
    port = 3306

    # 类的私有属性 - 不能被外部直接访问，保持内部状态
    __unit_info = 12345

    def __init__(self):

        self.ip = '127.0.0.1'
        self.user_account = 'root'
        self.__func_info = "private"
    
    def select(self):
        print("---select----")
        
    
    def update_info(self):
        # 类的私有属性，类的内部可以修改和访问
        print(f"before update {self.__unit_info}") # before update 12345
        MysqlConnectionPrivate._MysqlConnectionPrivate__unit_info = 54321
        print(f"after update {self.__unit_info}") # after update 54321

        # 实例的私有属性，类的内部可以修改和访问
        print("--before:", vars(self)) # --before: {'ip': '127.0.0.1', 'user_account': 'root', '_MysqlConnectionPrivate__func_info': 'private'}
        self.__func_info = "private_updated"
        print("--after:", vars(self)) # --after: {'ip': '127.0.0.1', 'user_account': 'root', '_MysqlConnectionPrivate__func_info': 'private_updated'}

    
db_p = MysqlConnectionPrivate()

db_p.update_info()

# 类的私有不可以被实例修改和访问
# db_p.__unit_info # AttributeError: 'MysqlConnectionPrivate' object has no attribute '__unit_info'
print(f"before update {vars(db_p)}")
db_p.__unit_info = 3333
print(f"after update {vars(db_p)}")

# 类的私有方法可以通过名称修饰的方式访问和修改
print(f"before update {MysqlConnectionPrivate._MysqlConnectionPrivate__unit_info}")
MysqlConnectionPrivate._MysqlConnectionPrivate__unit_info = 2333
print(f"after update {MysqlConnectionPrivate._MysqlConnectionPrivate__unit_info}")

"""
Python 没有真正意义上的 private。

双下划线(__)实现的是 Name Mangling（名称修饰），
目的是避免误用和子类命名冲突，
而不是彻底禁止访问。
"""

# 实例的私有属性不能实例直接访问和修改
#  print(db_p.__func_info) # AttributeError: 'MysqlConnectionPrivate' object has no attribute '__func_info'
print(vars(db_p)) # {'ip': '127.0.0.1', 'user_account': 'root', '_MysqlConnectionPrivate__func_info': 'private_updated', '__unit_info': 3333}
db_p.__func_info = 3222
print(vars(db_p)) # {'ip': '127.0.0.1', 'user_account': 'root', '_MysqlConnectionPrivate__func_info': 'private_updated', '__unit_info': 3333, '__func_info': 3222}

# 实例的私有属性可以通过名称修饰的方式访问和修改
print(
    db_p._MysqlConnectionPrivate__func_info   # private_updated
)

db_p._MysqlConnectionPrivate__func_info = "private_updated_again"

print(
    db_p._MysqlConnectionPrivate__func_info # private_updated_again
)

# 实例的私有方法不能通过类直接访问，但是可以通过类的方法访问
# print(MysqlConnectionPrivate._MysqlConnectionPrivate__func_info) # AttributeError: type object 'MysqlConnectionPrivate' has no attribute '_MysqlConnectionPrivate__func_info'
```

```python
# 类方法
class MysqlConn():
    """
    类方法的总结与练习
    """

    pwd = 123456
    port = 3306
    name = "ccc"

    def __init__(self):

        self.ip = '127.0.0.1'
        self.user_account = 'root'
    
    def select(self):
        print("---select----")
    
    @classmethod
    def ccc_info(cls):
        cls.name = "className"
        print("this is classmethod")
        return ""


# 类方法绑定在类上而非实例
MysqlConn.ccc_info() # this is classmethod - 可以不用实例化直接调用
# MysqlConn.select() # TypeError: MysqlConn.select() missing 1 required positional argument: 'self' 

conn = MysqlConn()

# 类方法改变类属性的值影响的是类的属性，不会影响实例的属性
print(conn.name) # 要注释掉前面的  MysqlConn.ccc_info() 否则类属性 name就已经被改了
print(conn.ccc_info())

# 类方法可以通过实例调用
conn.ccc_info()

# 类方法改变类属性的值影响的是类的属性，不会影响实例的属性
print(conn.name)
print(conn.ccc_info())
```

创建一个学生信息管理系统，能够添加、获取和更新学生信息。

要求：

创建一个名为Student的类，包含以下属性和方法：
类属性：total_students，记录学生总数。
实例属性：name、age、grade。
类私有属性：__school_name，设置为"XYZ学校"。
实例私有属性：__id，为学生ID。
构造方法：接受name、age、grade参数，并自动生成ID。
实例方法：get_info，返回学生信息。
类方法：get_school_name，返回学校名称。
静态方法：is_adult，接受age参数，判断是否为成年人。
在主程序中创建几个Student对象，演示添加、获取和更新学生信息的功能。

```python
class Student:

    total_students = 0
    __school_name = "XYZ学校"

    def __init__(self,name,age,grade):
        self.name = name
        self.age = age
        self.grade = grade

        # 在类中修改类属性
        Student.total_students +=1
        
        # 实例私有属性无法被实例直接修改或访问，而每个对象是一个学生，每次添加一个学生都重新实例化，无法实现自增，所以借助类属实现更改
        self.__id = Student.total_students

    def get_info(self):
        return {
            "id":self.__id, # 实例的私有属性无法直接通过object.__id，但是可以通过类内部的方法作为公开入口来访问
            "name":self.name,
            "age":self.age,
            "grade":self.grade,
        }
    
    def update_info(self,new_age=None,new_grade=None):
        
        if new_age is not None:
            self.age = new_age
        
        if new_grade  is not None:
            self.grade = new_grade

    @classmethod
    def get_school_name(cls):
        return cls.__school_name # 类的私有封装在类方法里，通过类方法作为公开入口访问属性代替依赖名称修饰直接访问
    
    @staticmethod
    def is_adult(age): # 静态方法不能加self
        return age>=18

if __name__ == "__main__":
    stu1 = Student("Adam",18,"高三")
    stu2 = Student("Charli",16,"高一")

    print(stu1.get_info())
    print(stu2.get_info())


    print(Student.get_school_name())

    print(f"18岁是成年人。 - {Student.is_adult(18)}")
    print(f"16岁是成年人。 - {Student.is_adult(16)}")

    stu1.update_info(new_age=19,new_grade="大一")
    stu2.update_info(new_grade="高二")

    print(stu1.get_info())
    print(stu2.get_info())

    print(f"目前学校共有学生 {Student.total_students} 个")
```
