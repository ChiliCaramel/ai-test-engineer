```python
import pymysql

db_connect = None
cursor = None

try:

    db_connect = pymysql.connect(
        host=host,
        user=user,
        password=password,
        port=port,
        database=database,
        charset='utf8mb4'
    )

    cursor = db_connect.cursor(
        pymysql.cursors.DictCursor
    )

    cursor.execute(
        "SELECT * FROM student"
    )

    result = cursor.fetchall()

    print(result)

except Exception as e:

    print("执行失败：", e)

    if db_connect:
        db_connect.rollback()

finally:

    if cursor:
        cursor.close()

    if db_connect:
        db_connect.close()

    print("资源释放完成")
```

```python
insert_sql_a = """
INSERT INTO student (id,Name,Sex,Birth,Departmen,Address) 
VALUES 
(112, '李四', '男', '2001-02-02', '英文系', '上海市徐汇区')
"""
try:
    # 执行插入语句
    cursor.execute(insert_sql_a)

    # 增删改要commit
    db_connect.commit()

    # 查询刚插入的数据
    select_sql_a = 'select * from student where id = 112 ;'

    # 执行查询语句
    cursor.execute(select_sql_a)

    # 获取数据
    res_a = cursor.fetchall()

    # 打印数据
    print(res_a)
except Exception as e:
    # 出现异常回滚
    db_connect.rollback()

    print("执行失败")
    print(e)

```

```python
update_sql_a = "update student set Address = '北京市东城区' where id = 112 "

try:
    # 执行更新语句
    cursor.execute(update_sql_a)
    
    # commit
    db_connect.commit()

    select_sql_aa = 'select Address , stu.* from student stu where id = 112'

    cursor.execute(select_sql_aa)

    res_aa = cursor.fetchall()
    print(res_aa)
except Exception as e:
    # 出现异常回滚
    db_connect.rollback()

    print("执行失败")
    print(e)
```

```python
select_sql_aaa = 'select * from student where id = 111'
delete_sql_a = 'delete from student where id = 111'

try:
    # before delete
    cursor.execute(select_sql_aaa)
    res_aaa = cursor.fetchall()
    print(f"before delete:{res_aaa}")

    # delete
    cursor.execute(delete_sql_a)
    db_connect.commit()

    # after delete
    select_sql_aaa = 'select * from student where id = 111'
    cursor.execute(select_sql_aaa)
    res_aaa_after = cursor.fetchall()
    print(f"after executed delete query:{res_aaa_after}")

except Exception as e:
    # 出现异常回滚
    db_connect.rollback()

    print("执行失败")
    print(e)
```

```python
# update 练习
update_query = """
update student set Departmen = '计算机系' , Address = '上海市浦东新区' where id = 201
"""

try:
    cursor.execute(update_query)
    db_connect.commit()

    new_update_sel = 'select * from student where id = 201'
    cursor.execute(new_update_sel)

    res_update = cursor.fetchall()
    print(res_update)

except Exception as e:
    # 出现异常回滚
    db_connect.rollback()

    print("执行失败")
    print(e)
```

```python
select_all_computer = "select * from student where Departmen = '计算机系'"
try:
    cursor.execute(select_all_computer)

    res_computer = cursor.fetchall()
    count_com = len(res_computer)

    print(res_computer)
    print(f"计算机系共{count_com}名学生")
except Exception as e:
    # 出现异常回滚
    db_connect.rollback()

    print("执行失败")
    print(e)
```

```python
# 删除练习
delete_new_stu = "delete from student where id = 201"
select_to_verify = "select * from student where id = 201"

try:
    cursor.execute(delete_new_stu)
    db_connect.commit()

    cursor.execute(select_to_verify)

    res_to_verify = cursor.fetchone()

    if not res_to_verify:
        print("学生已删除")
    else:
        print("删除失败")
except Exception as e:
    # 出现异常回滚
    db_connect.rollback()

    print("执行失败")
    print(e)
finally:
    # 最后一条语句，加一个finally 直接关掉cursor 和 connection
    if cursor:
        cursor.close()
    
    if db_connect:
        db_connect.close()
    
    print("Rlease DB 资源")
```

```python
# 1. 用pymysql 连接上你之前配置的数据库，接下来的所有操作都是用pymysql

import pymysql

# 指定数据库连接信息

host = '124.223.193.254' # 数据库ip地址
user = 'root' # 连接数据库的用户名
port = 3306 # 连接数据库的端口号
password = 'Cara@8888' #连接数据库的密码
database = 'mysql' #连接的是哪个数据库
charset = 'utf8' # 编码格式

# 连接数据库
db_connect = pymysql.connect(host=host,user=user,port=port,password=password,database=database,charset=charset)

# 创建一个可执行sql的光标
cursor = db_connect.cursor(cursor=pymysql.cursors.DictCursor)
```

```python
# 2. 查询student表中的所有数据，要求用 [{'id':'1', 'name':'校长'},{'id':'1', 'name':'小张'},{'id':'2', 'name':'小明'}]的格式打印查询出来的所有数据

select_all_stu = 'select * from student'

try:
    cursor.execute(select_all_stu)
    rows = list(cursor.fetchall())

    res_all_stu = []
    
    for row in rows:
        res_all_stu.append(
            {
                "id":row["id"],
                "name":row["Name"]
            }
        )
    print(res_all_stu)
except Exception as e:
    print("执行失败")
    print(e)
```

```python
# 3. 在第二点的基础上，把所有人的姓名打印出来，用 ['小张','小明']的格式打印
name_list = []

for stu in res_all_stu:
    name_list.append(stu["name"])

print(name_list)
```

```python
# 4. 用pymysql实现在student表中插入数据

insert_stu = """
INSERT INTO student (id,Name,Sex,Birth,Departmen,Address) 
VALUES 
(203, 'AAA', '男', '2001-02-02', '英文系', '上海市徐汇区')
"""

try:
    cursor.execute(insert_stu)
    db_connect.commit()

except Exception as e:
    db_connect.rollback()

    print("执行失败")
    print(e)
```

```python
# 5. 用pymysql实现在student表中删除数据

delete_stu = """
delete from student where id = 202
"""

try:
    cursor.execute(delete_stu)
    db_connect.commit()

except Exception as e:
    db_connect.rollback()
    print("执行失败")
    print(e)
```

```python
# 6. 用pymysql实现在student表中更新数据

update_stu = """
update student set Name = 'LiSi' where Name = '李华'
"""

try:
    cursor.execute(update_stu)
    db_connect.commit()

except Exception as e:
    db_connect.rollback()
    print("执行失败")
    print(e)

finally:
    cursor.close()
    db_connect.close()
```
