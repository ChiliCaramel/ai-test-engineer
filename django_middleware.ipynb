中间件作用：
1. 处理请求和响应的机制，允许在Django框架的请求/响应处理过程中介入，并执行各种操作。
2. 中间件是一个类
3. 中间件可以用于执行各种任务，例如修改请求或响应，记录日志、身份验证等。

其他：
请求处理：在请求到达views之前，中间件可以对请求进行预处理。 比如身份验证、url重写、添加或修改请求头等。
响应处理：在views生成响应之后，中间件可以对响应进行处理。包括添加或修改请求头、压缩响应内容、重定向等。
异常处理：中间件可以捕获在请求处理过程中发生的异常，然后执行相关的操作，比如记录日志、发送警报或自定义错误页面。
安全性检查：中间件可以执行各种安全性检查，如防止跨站脚本攻击(xss)、跨站请求伪造（CSRF）保护等。
会话管理：中间件可以处理会话相关的任务，比如创建、更新、删除会话数据。


Django中间件内置魔法方法__call__ - 允许类的实例像函数一样被调用

for example:


```python
class CallableClass:
    def __init__(self,name):
        self.name = name

    def __call__(self,*args,**kwargs):
        print(f"Instance {self.name} is called with arguments:{args} and keybword arguments:{kwargs}")

# 创建实例
obj = CallableClass("example")

# 调用实例，无需 obj.xxx 
obj(1,2,key=3)
```
