Django
1. Pyhton最流行的web框架
2. 能做什么：
    - 开发网站后端
    - 提供API接口（给前端或移动端调用）
    - 管理数据库
    - 用户认证（登录、注册、权限）
    - 自带后台管理系统
3。 核心组件 - MTV 架构
    Model - 数据库模型，定义数据结构
    Template - 模版，负责页面展示（HTML）
    View - 视图，处理业务逻辑
4. MTV交互顺序 ：
User发送请求 - URLs路由 - View处理 - Model查数据 - Template渲染 - 返回响应
5. Django - ORM - Object Relational Mapping 对象关系映射
   作用：将数据库的表映射为python的类
        将数据库的行映射成python的对象
        不用写SQL，直接用python操作数据库
    ex:
    # 非Django 方式：使用pymysql库，手写SQL
    cursor.execute("select * from Student where name = 'zhangsan'")

    # Django - ORM
    Student.objects.filter(name = "zhangdan")

除此之外，ORM防止了SQL注入（因为不用写query）
且当项目需要换数据库时，只需更改配置，不必大量更换代码。

6. Django的安全性：
   CSRF token - 防伪造请求 - 原理：在表单中加入token验证
   自动转义 - 防XSS脚本注入 - 原理：Django Template将user输入的HTML标签（因为通过API获取，所以都会填充在{}中，触发template自动转义）自动转译成文字，而不是执行脚本。
   ORM - 防SQL注入

Django 配置数据库
1. 在settings.py - DATABASES 中配置
2. 注释掉默认的 db.sqlite3 相关的配置 , 换成自己的配置
DATABASES = {
    'default': {
        # 'ENGINE': 'django.db.backends.sqlite3',
        # 'NAME': os.path.join(BASE_DIR, 'db.sqlite3'),
        'ENGINE':'django.db.backends.mysql',
        'NAME':'mysql'
    }
3. 创建表的完整流程：
   ① models.py 定义表结构
   ② python manage.py makemigrations  → 生成迁移文件
   ③ python manage.py migrate         → 同步到数据库
   缺少任何一步都不会建表！

额外补充知识点：
在 Django 中，当你创建一个模型时，如果没有明确指定一个主键字段，Django 会自动为你添加一个名为 id 的主键字段。
这个字段是一个自增的整数类型字段，用于唯一标识数据库中的每条记录。

from django.db import models

class MyModel(models.Model):
    my_field = models.CharField(max_length=100)

尽管我们没有在 MyModel 中定义主键，Django 会在幕后自动添加一个 id 字段作为主键。在数据库中，这个字段通常会被创建为一个自增的整数类型字段。
